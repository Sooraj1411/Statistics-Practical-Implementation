# Measure of Dispersion
#### 1. Variance
#### 2. Standard Deviation

```python
print(stat.variance(AGES))               #statistics.variance formula sample variance ka formula use krta h (n-1) wala

print(stat.pvariance(AGES))              #statistics.pvariance formula population variance ka formula use krta h (n) wala

print(np.var(AGES))                      #numpy.var formula population variance ka formula use krta h (n) wala
```

```python
def vari(data):
    n = len(data)
    mean = sum(data) / n
    var = 0
    for i in data:  # Iterate over the elements in data
        var += (i - mean) ** 2  # Calculate variance component for each element
    var = var / n  # Divide by the number of elements to get variance
    return var

# Example usage
print(vari(AGES))
```

```python
def vari(data , dof = 0):
    n = len(data)
    mean = sum(data) / n
    var = 0
    for i in data:  # Iterate over the elements in data
        var += (i - mean) ** 2  # Calculate variance component for each element
    var = var / (n-dof)  # Divide by the number of elements to get variance
    return var

# Example usage
print(vari(AGES,1))
```

```python
import math
print(math.sqrt(np.var(AGES)))
print(np.std(AGES))
```

# Checking whether dist is normal or not

```python
import matplotlib.pyplot as plt
import seaborn as sns  # Import seaborn for sns.histplot
import scipy.stats as stat
import pylab

def plot_data(sample):
    plt.figure(figsize=(10,6))
    
    # Histogram plot
    plt.subplot(1,2,1)
    sns.histplot(sample, kde = True)
    
    # Q-Q plot
    plt.subplot(1,2,2)
    stat.probplot(sample, dist='norm', plot=pylab)
    
    # Show the plots
    plt.show()

# Example usage
import numpy as np
sample = np.random.normal(0, 1, 100)  # Generate a normal distribution sample
plot_data(sample)
```
